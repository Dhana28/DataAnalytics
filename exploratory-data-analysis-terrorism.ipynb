```python
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import pandas as pd

# Sample data (replace this with your actual data)
data = {
    'Year': [1982, 1983, 1984, 1985, 1986, 1987, 1988],
    'Attack Type': ['Bombing', 'Hostage Taking', 'Bombing', 'Hostage Taking', 'Bombing', 'Hostage Taking', 'Bombing'],
    'Count': [10, 15, 8, 12, 7, 9, 6]
}

df = pd.DataFrame(data)

# Create a scatter plot for individual attacks
scatter_data = []
attack_types = df['Attack Type'].unique()

for attack_type in attack_types:
    subset = df[df['Attack Type'] == attack_type]
    scatter_trace = go.Scatter(
        x=subset['Year'],
        y=subset['Count'],
        mode='markers',
        name=attack_type,
        marker=dict(size=10)
    )
    scatter_data.append(scatter_trace)

# Create a line plot for trend
trend_trace = go.Scatter(
    x=df['Year'],
    y=df.groupby('Year')['Count'].sum(),
    mode='lines+markers',
    name='Total Attacks',
    line=dict(shape='spline')
)

data = scatter_data + [trend_trace]

# Create annotations
annotations = []
for _, row in df.iterrows():
    annotation = go.layout.Annotation(
        x=row['Year'],
        y=row['Count'],
        xref='x',
        yref='y',
        text=f"{row['Attack Type']}<br>{row['Count']} attacks",
        showarrow=True,
        arrowhead=2,
        ax=0,
        ay=-40
    )
    annotations.append(annotation)

layout = go.Layout(
    title='Terror Attack Trends from 1982 to 1988',
    xaxis=dict(title='Year', nticks=7),
    yaxis=dict(title='Number of Attacks'),
    annotations=annotations,
    hovermode='closest'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='complex_representation', show_link=False)
```

**The Above map looks untidy even though it can be zoomed in to view the Country in question. Hence in the next chart, I have used Folium's Marker Cluster to cluster these icons. This makes it visually pleasing and highly interactive.**

```python
import folium
from folium.plugins import MarkerCluster

# Assuming you have a DataFrame named terror_df_group with latitude, longitude, Group, and Country columns

m1 = folium.Map(location=[20, 0], tiles="CartoDB positron", zoom_start=2)
marker_cluster = MarkerCluster(
    name='clustered icons',
    overlay=True,
    control=False,
    icon_create_function=None
)

for i in range(0, len(terror_df_group)):
    marker = folium.Marker([terror_df_group.iloc[i]['latitude'], terror_df_group.iloc[i]['longitude']]) 
    popup = 'Group: {}<br>Country: {}'.format(terror_df_group.iloc[i]['Group'], terror_df_group.iloc[i]['Country'])
    folium.Popup(popup).add_to(marker)
    marker_cluster.add_child(marker)

marker_cluster.add_to(m1)
folium.TileLayer('openstreetmap').add_to(m1)
folium.TileLayer('CartoDB positron').add_to(m1)  # Use the correct name for the CartoDB positron layer
folium.TileLayer('cartodbdark_matter').add_to(m1)
folium.TileLayer('stamentoner').add_to(m1)
folium.LayerControl().add_to(m1)

m1
```

```python
# Total Number of people killed in terror attack
killData = terror.loc[:,'Killed']
print('Number of people killed by terror attack:', int(sum(killData.dropna())))# drop the NaN values
```

```python
# Let's look at what types of attacks these deaths were made of.
attackData = terror.loc[:,'AttackType']
# attackData
typeKillData = pd.concat([attackData, killData], axis=1)
```

```python
typeKillData.head()
```

```python
typeKillFormatData = typeKillData.pivot_table(columns='AttackType', values='Killed', aggfunc='sum')
typeKillFormatData
```

```python
typeKillFormatData.info()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have the required data
labels = typeKillFormatData.columns.tolist()  # Convert the DataFrame columns to a list
transposed = typeKillFormatData.T  # Transpose the DataFrame
values = transposed.values.tolist()  # Convert the transposed DataFrame to a list of lists

# Create a stacked bar chart
fig, ax = plt.subplots(figsize=(10, 15))

# Plot each category as a stacked bar
bottom = np.zeros(len(labels))
for row in values:
    ax.bar(labels, row, bottom=bottom)
    bottom += row

ax.set_title('Types of terrorist attacks that cause deaths')
ax.set_xlabel('Attack Types')
ax.set_ylabel('Number of Deaths')
ax.legend(labels, loc='upper right', fontsize=10, bbox_to_anchor=(1.25, 1.25))
plt.xticks(rotation=65, ha='right')
plt.tight_layout()

plt.show()
```

**<center>Armed assault and bombing/explosion are seen to be the cause of 77% of the deaths in these attacks. This rate is why these attacks are used so many times in terrorist actions. This is how dangerous weapons and explosives are to the world.</center>**
